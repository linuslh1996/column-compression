```python
import pandas as pd
import matplotlib.pyplot as plt

dictionary = pd.read_csv("sizes_none.txt", sep="|")
dict_size = dictionary["size_in_bytes"].sum()
print("dictionary size", dict_size)

for name in [("fsba", "FixedSize"), ("SIMDBP", "SimdBp128"), ("turbopfor", "TurboPForBitpacking")]:
    runtimes = pd.read_csv("sizes_" + name[0] + ".txt", sep="|")
    s = runtimes["size_in_bytes"].sum()

    #print(runtimes)


    #cols_with_name_flags = simdcai["encoding_type"].str.match(".*" + name + ".*")
    cols_with_name_flags = runtimes["vector_compression_type"].str.match(".*" + name[1] + ".*")
    cols_with_name = runtimes[cols_with_name_flags]    

    cols_with_name_dict_encoded = dictionary[cols_with_name_flags]

    size1 = cols_with_name["size_in_bytes"].sum()
    size2 = cols_with_name_dict_encoded["size_in_bytes"].sum()


    print(name, "size:", s)
    print("\ttotal percentage:", s / dict_size, "\n\tencoded col percentage:", size2 / dict_size, "\n\tencoded col compression:", size1 / size2)



```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(30,15)})


d = dictionary[["    column_name",'table_name','size_in_bytes']]
d['id'] = d["    column_name"].map(str) + d.table_name
#print(d)
#print(d.table_name)

d = d.groupby("id")["size_in_bytes"].sum()

#print(d["size_in_bytes"].sum())
d.plot.barh()
```
